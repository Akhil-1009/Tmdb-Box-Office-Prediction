1. CHANGE THE YEAR COLOUM INTO DATE DAY AND MONTH

```python
import pandas as pd

# Load the dataset
file_path = '/content/normalised_train_tmdb.csv'
df = pd.read_csv(file_path)

# Convert release_date (string) to release_day, release_month, release_year (separate columns, integer format)
df['release_date'] = pd.to_datetime(df['release_date'])

df['release_day'] = df['release_date'].dt.day
df['release_month'] = df['release_date'].dt.month
df['release_year'] = df['release_date'].dt.year

# Drop the original release_date column
df.drop('release_date', axis=1, inplace=True)

```

```python
print("\nNew columns after transformation:")
print(df[['release_day', 'release_month', 'release_year']])
```

HOT ENCODING FOR THE GENRE COLOUM

```python
df['genres']=df['genres'].apply(lambda x:x.split(', '))
genres=set()
for genre in df['genres']:
    genres.update(genre)
for genre in genres:
    df[genre]=df['genres'].apply(lambda x: 1 if genre in x else 0)
# drop the genre coloum
df.drop('genres', axis=1, inplace=True)
df.head()
```

3. HOT ENCODING FOR PRODUCTION COUNTRIES

```python
df['production_countries']=df['production_countries'].apply(lambda x:x.split(', '))
countries=set()
for country in df['production_countries']:
    countries.update(country)
for country in countries:
    df[country]=df['production_countries'].apply(lambda x: 1 if country in x else 0)
# drop the coloum
df.drop('production_countries', axis=1, inplace=True)
df.head()
```

```python
df.columns
```

FOR SPOKEN LANGUAGES

```python
df['spoken_languages']=df['spoken_languages'].apply(lambda x:x.split(', '))
languages=set()
for language in df['spoken_languages']:
    languages.update(language)
for language in  languages:
    df[language]=df['spoken_languages'].apply(lambda x: 1 if language in x else 0)
# drop the coloum
df.drop('spoken_languages', axis=1, inplace=True)
df.head()
```

CONVERT CATEGORICAL VALUES TO NUMERICAL VALUES

```python
# Convert the status column to numerical value 1 = "Released", 0 = "Not Released"
df['status'] = df['status'].apply(lambda x: 1 if x == 'Released' else 0)
```
